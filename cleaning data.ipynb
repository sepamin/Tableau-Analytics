```python
# dependencies
import random
import pandas as pd
import requests
import json
import numpy as np
import glob
import json
```

```python
# combine all citi bikeshare csv into one mega csv

path = "data"
all_files = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
timeYr = []

for file_ in all_files:
    df = pd.read_csv(file_,index_col=None, header=0)
    timeYr.append(df)
frame = pd.concat(timeYr)
frame.to_csv("data/data_merged.csv", index=False)
```
